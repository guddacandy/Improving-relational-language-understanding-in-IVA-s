**Loading the Dataset.**

```python
#Loading the rsics_dataset and converting it from tar.gz to csv file format.
import tarfile
import csv

#open the rsics_dataset.tar.gz output_file
with tarfile.open('rsics_dataset.tar.gz', 'r:gz') as tar:
  #extract and process each file
  for member in tar.getmembers():
    #checking for any csv files in the dataset and processing accordingly if not
    if member.name.endswith('.csv'):
      #Extract the file into a csv format
      file = tar.extractfile(member)
      #Read the data in the csv file
      csv_data = file.read().decode('utf-8') #Assuimg that the possible text in the data is encoded in utf-08
      #Extract the csv data to a new csv file
      with open ('rsics.csv', 'a') as rsics_file:
        rsics_file.write(csv_data)

print('converted rsics file')


```
